```python
###Representation of the LDOS in a crystalline structure

import pandas
import math
import numpy as np

data=pandas.read_table("output_0.1_wo_exc_power1.dat", delimiter=r"\s+")
LDOSx=data.iloc[:,0]
LDOSy=data.iloc[:,1]
LDOSz=data.iloc[:,2]
LDOS=(LDOSx+LDOSy+LDOSz)/3
Xmin = 1.5
Xmax = 12
step = len(data)
X=np.linspace(Xmin,Xmax-((Xmax-Xmin)/step),(step))
Xf=[]
LDOSf=[]
for i in range(len(LDOS)):
    LDOSf.append(LDOS[i])
    
%pylab inline
import matplotlib.pyplot as plt
rm=2*1.14707;

plt.plot(rm/X,LDOSf,label='LDOS')
plt.scatter(rm/4.923047,7,label='$r_m$/$\lambda$=0.466')
plt.scatter(0.795,0.1,label='$r_m$/$\lambda$=0.795',c='g')
plt.scatter(0.872,1,label='$r_m$/$\lambda$=0.872',c='r')
plt.xlabel('$r_m$/$\lambda$',fontsize=20)
plt.ylabel('$\Gamma$/$\Gamma_0$',fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
plt.legend(fontsize=15)
plt.ylim((0,9))
plt.xlim((0.,1.5))
plt.fig = matplotlib.pyplot.gcf()
plt.fig.set_size_inches(12, 8)
```
